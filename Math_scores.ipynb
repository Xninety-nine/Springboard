```python

plt.figure(figsize=(12, 6))


sns.histplot(data=df, bins=20, kde=True)

# Customize the plot
plt.title('Distribution of Math Scores ', fontsize=16) #Adjust fontsize and title 
plt.xlabel('Math Scores', fontsize=12) #Adds label for X axis along with a fontsize of 12
plt.ylabel('Frequency', fontsize=12)

# Adds mean age line that is red
mean_age = df[100].mean()
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Mean Age: {mean_age:.2f}') 

# Adds median age line that is green
median_age = df[100].median()
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=2, label=f'Median Age: {median_age:.2f}')

# Adds legend
plt.legend(fontsize=10)

# Shows grid
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
