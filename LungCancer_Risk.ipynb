## Observations: Possible BUG

Ordinal encoder encoded risk levels based on the alphabetical order of the name H-L-M. This is not good and might make it harder to interpret results during during clustering and modeling.

```python
risk_level = {'Low':0, 'Medium':1, 'High':2}

df_encoded['Level'] = df_encoded['Level'].map(risk_level)
```

# Clustering:

### Purpose: Find subtypes of patients or feature-based groupings.

## Question: Do clusters correspond to risk groups or known patterns?



```python
df_cluster = df_encoded.copy()
```

```python
from sklearn.preprocessing import StandardScaler

X = df_cluster.drop('Level', axis=1)


scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

## PCA:

Using Principal Component Analysis for dimensionality reduction. 

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) #tried 2, 11 and 3

X_pca = pca.fit_transform(X_scaled)
```

```python
#Calculating cumsum explained variance to determine optimal number of principal components
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
```

```python
# Elbow Method 
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.axhline(y=0.90, color='r', linestyle='--', label='90% variance')
plt.axhline(y=0.95, color='g', linestyle='--', label='95% variance')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of PCA Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Observations:

- PCA is linear and this is clearly non-linear. 

- Perhaps t-SNE would be better for this data set. 

```python
#Kmeans clustering
from sklearn.cluster import KMeans 

kmeans = KMeans(n_clusters=3, random_state= 42)
clusters = kmeans.fit_predict(X_pca)
```

```python
cluster_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])#,'PC3'])#,'PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11'])
cluster_df['Cluster'] = clusters
```

```python
#Visualization of clusters
plt.figure(figsize=(8, 6))
scatter = plt.scatter(cluster_df['PC1'], cluster_df['PC2'], c=cluster_df['Cluster'], cmap='viridis', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA + Clustering')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()
```

### Evaluate Clusters

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_pca, clusters)
print("Silhouette Score:", score)
```

### Observation: Curse of Dimensionality. 

- I first tried two principal components and got a silhouette score of 64%

- I then tried 11 Principal components which represented 90% of the variance in the dataset but silhouette score dropped to 34%.

- I also tried 3 principal components which explained 60% of variance in data and silhouette score was 53%

Silhouette Score is above 50% which is decent. Tuning is needed or perhaps trying different clustering Algorithms or perhaps clustering without using PCA.

```python
from sklearn.metrics import silhouette_score

scores = []
for k in range(2, 40):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, labels)
    scores.append((k, score))
    print(f"k = {k}, Silhouette Score = {score:.3f}")
```

```python
#Elbow to visualize best cluster number
k_vals, silhouette_vals = zip(*scores)

plt.plot(k_vals, silhouette_vals, marker='o')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

## Train_Test_Split

```python
key_predictors = ['Air Pollution', 'Smoking', 'Genetic Risk', 'Alcohol use',
    'Obesity', 'OccuPational Hazards', 'Dust Allergy']

X_reduced = df_encoded[key_predictors]
```

```python
import statsmodels.api as sm
from statsmodels.miscmodels.ordinal_model import OrderedModel
from sklearn.model_selection import train_test_split

# Define the features (X) and the target (y)
X = df_encoded.drop(columns=['Level'])
y = df_encoded['Level']

X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.25, random_state=22) #switched X for X_reduce to avoid multicolinearity 


```

## Model: Here I chose an Ordinal Logistic Regression model, however I will also try Random Forest, Multinominal Logistic Regression and gradient boosting 

Reasoning: OLR maintains the integrity of the order. Since the Level are split into 'Low','Medium' and 'High' this order must be taken into account differences between these three Levels of cancer.

```python
# Fit the ordinal logistic regression model
model = OrderedModel(y_train, X_train, distr='logit')
result = model.fit(method='lbfgs', maxiter=100, disp=False)

# Print summary
print(result.summary())

# Predict on the test set
y_pred = result.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_pred_prob = result.predict(X_test) #Classification metrics can't handel a mix of multiclass and continuous-multioutput targets, need to convert prob into class labels 


y_pred = np.argmax(y_pred_prob, axis=1)
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")
```

## Model Performance Reflection:

Model was biased. It favored misclassifying categories as 'Medium'. After adding these arguments the Accuracy jumped from 38% to 84.4% (method='lbfgs', maxiter=100, disp=False).

## Hyperparameter Tunning on Logistic Regression Model

```python
from scipy.stats import uniform
```

```python
# Hyperparameter grid
param_distributions = {
    'C': uniform(0.01, 100),  # C values between 0.01 and 100
    'penalty': ['l2'],  # Only L2 regularization since 'lbfgs' solver does not support L1
    'solver': ['lbfgs'],  # Solver choice
    'max_iter': [100, 200, 300]  # Number of iterations for convergence
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Initialize the RandomizedSearchCV with Logistic Regression
random_search = RandomizedSearchCV(
    LogisticRegression(multi_class='multinomial', random_state=22),
    param_distributions=param_distributions,
    n_iter=50,  # Number of random combinations to try
    scoring='accuracy',  # Optimize for accuracy
    cv=5,  # 5-fold cross-validation
    verbose=1,
    random_state=22,  # Ensure reproducibility
    n_jobs=-1  # Use all available cores
)
```

```python
# Fit Randomized Search on training data

random_search.fit(X_train, y_train)
```

```python
#Print best parameters

print('Best Parameters', random_search.best_params_)

#Print best cross-val score

print('Best Cross-Validation Score:', random_search.best_score_)
```
